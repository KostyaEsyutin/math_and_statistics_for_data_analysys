```python
from itertools import combinations, permutations
from math import factorial
```

# Permutations

Asume, we are school managers and we want to schedule 4 subjects - math, physics, chemistry and biology. <br>
Some teachers want to start their woorkday earlier, some later. We don't know surely for now but we can suggest them all possible options. <br>
Firstly, we want to understand how much combinations we have to enumerate because we want to estimate which is better - to do this mannualy (30 sec per iteration) or automate (60 minute once and 3 sec per iteration)

```python
subjects = ["math", "physics", "chemistry", "biology"]

def print_possible_permutations(available_subjects, print_all: bool=False):
    print('available subjects:', available_subjects)
    possible_orders = [x for x in permutations(available_subjects)]
    if print_all:
        print('possible orders:', possible_orders)
    print('possible orders count:', len(possible_orders), end='\n\n')
    return

# asume we have only maath and physics
available_subjects = ["math", "physics"]
print_possible_permutations(available_subjects, print_all=True)

# let's add chemistry
available_subjects.append("chemistry")
print_possible_permutations(available_subjects, print_all=True)

# let's use all subjects
available_subjects.append("biology")
print_possible_permutations(available_subjects, print_all=True)
```

As we can see, the number of permutations increases fastly and we can now try to find formula describing this number.

Let's look at all possible positions for math and other subjects:
- <span style="color: red"> math</span>, other, other, other
- other, <span style="color: red"> math</span>, other, other
- other, other, <span style="color: red"> math</span>, other
- other, other, other, <span style="color: red"> math</span>

Asume we've fixed the position of math and now we have to permutate only 3 subjects. Then let's show all possible positions of physics:
- <span style="color: grey"> math</span>, <span style="color: red"> physics</span>, other, other
- <span style="color: grey"> math</span>, other, <span style="color: red"> physics</span>, other
- <span style="color: grey"> math</span>, other, other, <span style="color: red"> physics</span>

The same for chemistry and after that biology has only one place to be put. <p>
Well, seems like we have 4 possible positions for math or any other subject, then 3 possible positions for residual subjects, then 2...1 <br>
Consequently, the number of permutations is $4*3*2*1$ or $4!$

The same formula for N elements:
$P_n = n! $


```python
# let's check the formula above by calculating the number of permutations for 5 subjects (asume we have one more subject, say "english")
available_subjects.append("english")
print_possible_permutations(available_subjects)
print('number of permutations for 5 subjects:', factorial(len(available_subjects)))

# let's add three more subjects and check the formula again
available_subjects.extend(["history", "geography", "art"])
print_possible_permutations(available_subjects)
print('number of permutations for 8 subjects:', factorial(len(available_subjects)))
```

Now we have 8 subjects but students can learn only 4 subjects per day. <br>
We should take into account this restriction and count all possible permutations now.

Following almost the same logic as discribed above we can conclude that we have 4 vacant positions. <br>
How much subjects can be put on the first place? - 8 <br>
Now we've fixed 1 place and 7 are left. Then the second place can be taken by 7 subjects. <br>
Continuing this logic, we'll reach the fourth place, which can be taken by 5 subjects.

Summarizing all this thoughts, the number of permutations of 4 subjects from 8 is:
$8*7*6*5 $

Or: <p>
$$
 \frac{8*7*...*1}{4*3*2*1} = \frac{8!}{4!} = 8*7*6*5
 $$

 And generalizing:
 $$
 P(n, r) = \frac{n!}{(n-r)!}
 $$
 , where r - is the number of taken subjects

```python
# let's count the number of permutations for 4 subjects out of 8 available subjects
print("count for 4 subjects:", len([x for x in permutations(available_subjects, 4)]))

# let's check the formula for permutations of 4 subjects out of 8 available subjects
print("formula for 4 subjects:", factorial(len(available_subjects)) // factorial(len(available_subjects) - 4))

# let's count the number of permutations for 5 subjects out of 8 available subjects
print("count for 5 subjects:", len([x for x in permutations(available_subjects, 5)]))

# let's check the formula for permutations of 5 subjects out of 8 available subjects
print("formula for 5 subjects:", factorial(len(available_subjects)) // factorial(len(available_subjects) - 5))
```

# Combinations

Ok, we have 4 subjects per day which means that we have to ask 4 teachers to come to our school or students need to take 4 books. <br>
There is no difference between the same but reordered set of subjects. So, it is necessary to exclude repeated sets from our permutations.

To simplify, let's turn back to the initial set of subjects. Asume we've already got P(4,2) permutations. How much of them are repeated?

All possible permutations:
- ('math', 'physics') ('physics', 'math')
- ('math', 'chemistry') ('chemistry', 'math')
- ('math', 'biology') ('biology', 'math') 
- ('physics', 'chemistry') ('chemistry', 'physics')
- ('physics', 'biology') ('biology', 'physics')
- ('chemistry', 'biology') ('biology', 'chemistry')

We see that every permutation is repeated twice. Here is tricky but intuitive logic. <br>
We have to choose 2 subjects from 4. Asume we've got math and physics. How much permutations on this set? - 2*1. <br>
The same for every pair above.

Generalizing this discussion, we achive extention of permutation with restrictions formula which is called combinations:
$$
C_n^r = \frac{P(n, r)}{r!} = \frac{n!}{(n-r)!r!}
$$

```python
# let's count the number of combinations for 4 subjects out of 8 available subjects
print("count for combinations of 4 subjects:", len([x for x in combinations(available_subjects, 4)]))
# let's check the formula for combinations of 4 subjects out of 8 available subjects
print("formula for combinations of 4 subjects:", factorial(len(available_subjects)) // (factorial(4) * factorial(len(available_subjects) - 4)))
```
